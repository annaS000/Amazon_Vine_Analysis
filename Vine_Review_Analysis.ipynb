```python
import pandas as pd
```

```python
#read vine csv
df = pd.read_csv('/Users/annasand/Desktop/data/repos/Amazon_Vine_Analysis/resources/vine_table.csv')  
```

```python
#get reviews with total votes > 20
df = df.loc[df['total_votes'] > 19]
df
```

```python
#retrieve all the rows where the number of helpful_votes 
#divided by total_votes is equal to or greater than 50%
df = df.loc[df['helpful_votes']/df['total_votes']>=0.5]
df
```

```python
#total reviews
review_total = len(df['review_id'])
review_total
```

```python
#vine program paid df
paid_df = df.loc[(df['vine'] == 'Y')]
paid_df.head()
```

```python
#total and percent for paid
total_paid = len(paid_df['review_id'])
paid_percent = (total_paid*100/review_total)
print(f'Total Reviews (paid): {total_paid}')
print(f'Percent of Paid Reviews: {round(paid_percent, 2)}%')
```

```python
#vine program unpaid df
unpaid_df = df.loc[(df['vine'] == 'N')]
unpaid_df.head()
```

```python
#total and percent for unpaid
total_unpaid = len(unpaid_df['review_id'])
unpaid_percent = (total_unpaid*100/review_total)
print(f'Total Reviews (unpaid): {total_unpaid}')
print(f'Percent of Unpaid Reviews: {round(unpaid_percent, 2)}%')
```

```python
#vine program 5 stars df
fivestar_df = df.loc[df['star_rating'] == 5]
fivestar_df.head()
```

```python
fivestar_total = len(fivestar_df['review_id'])
fivestar_total
```

```python
#vine program 5 stars df (paid)
paid_fivestar_df = paid_df.loc[paid_df['star_rating'] == 5]
paid_fivestar_df.head()
```

```python
#total and percent 5 star (paid)
total5_paid = len(paid_fivestar_df['review_id'])
paid5_percent = (total5_paid*100/fivestar_total)
print(f'Total 5-Star Reviews (paid): {total5_paid}')
print(f'Percent of Paid 5-Star Reviews: {round(paid5_percent, 2)}%')
```

```python
#vine program 5 stars df (unpaid)
unpaid_fivestar_df = unpaid_df.loc[unpaid_df['star_rating'] == 5]
unpaid_fivestar_df.head()
```

```python
#total and percent 5 star (unpaid)
total5_unpaid = len(unpaid_fivestar_df['review_id'])
unpaid5_percent = (total5_unpaid*100/fivestar_total)
print(f'Total 5-Star Reviews (unpaid): {total5_unpaid}')
print(f'Percent of Unpaid 5-Star Reviews: {round(unpaid5_percent, 2)}%')
```
